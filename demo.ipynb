```python
import pandas as pd
import torch

from REDEvaluator import REDEvaluator

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
# REDEvaluator args
data_path = "./data/RED6k-toy.jsonl"

evaluator = REDEvaluator(
    model_name="google/gemma-3-1b-it", # local path or huggingface model path
    input_column="question", # Match to your data
    document_column="contexts", # Match to your data
    metric="query completeness", # Check prompts.py for starter metrics
    openai_api_key="YOUR_OPENAI_KEY", # OpenAI API key
    device=device, # Device to run the model on
)

df = pd.read_json(data_path, lines=True) # load data

results = evaluator.process_and_evaluate(df)
temp_df = pd.DataFrame(results)
temp_df.to_json("./eval_results/toy_results.jsonl", orient="records", lines=True)
temp_df.head()
```
